# Unsupervised model

```python
from gensim.models import LdaModel

lda_model = LdaModel.load('../models/lda/lda_model.gensim')

topics = lda_model.show_topics(num_topics=10, num_words=10, formatted=False)

for topic_id, word_probs in topics:
    print(f'\n➡️ Topic {topic_id}:')
    for word, prob in word_probs:
        print(f'   - {word:<15} {prob:.4f}')
```
