```python
from datetime import datetime
from stackapi import StackAPI

import pandas as pd
```

```python
SITE = StackAPI('stackoverflow')
SITE.page_size = 50   
SITE.max_pages = 1     


from_date = int(datetime(2021, 3, 1).timestamp())
to_date = int(datetime(2025, 3, 1).timestamp())


response = SITE.fetch('questions',
                      fromdate=from_date,
                      todate=to_date,
                      min=51,
                      tagged='python',
                      sort='votes')


items = response['items']


data = [{
    'date': datetime.fromtimestamp(q['creation_date']),
    'title': q['title'],
    'tags': q['tags'],
    'score': q['score']
} for q in items]
```

```python
df = pd.DataFrame(data)
df
```
